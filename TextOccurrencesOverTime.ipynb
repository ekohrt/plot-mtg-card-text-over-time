# Occurrences of keywords/phrases in Magic card text over time

We show how to search through all cards in Magic: the Gathering for occurrences of some keyword or expression in the card text. Then we plot the number of newly printed cards each year that contain the given expression. These graphs can illuminate some interesting trends in the history of the game, as new game mechanics are invented and others are phased out.

```python
import re
from matplotlib import pyplot as plt
from TextOverTime import searchCardsForMatches, getAllEarliestDates, plotDates
```

### Find all cards that contain some text

Write a regular expression, then search through all cards looking for matches in the card text (lowercase only).
We use a file from [MTGJSON](https://mtgjson.com/file-models/card-atomic/) to search for cards.

```python
regex = r'(.*)regenerate(.*)'
pattern = re.compile(regex)
matches = searchCardsForMatches(pattern)
print(matches[:10], "...")
```

### Get the earliest release date for each card

We are looking at only the earliest printing of each card, so the resulting graphs will not show reprints.

```python
allDates = getAllEarliestDates(matches)
print(allDates[:10], "...")
```

### Plot the data

Let's plot the use of the keyword ***'regenerate'*** over Magic's history. The mechanic has been around since the very beginning in 1993. However, Wizards of the Coast has stated that they are [phasing out the "regenerate" ability](https://tappedout.net/mtg-forum/spoilers-rumors-and-speculation/wotc-is-phasing-out-regeneration/), so the graph should demonstrate a downward trend with no new ***'regenerate'*** cards printed since 2016:

```python
plotDates(allDates, regex)
```

### Other Examples

```python
# The ability "Double Strike" was introduced in the "Legions" set from 2003, and has increased in popularity.
regex2 = r'(.*)double strike(.*)'
allDates2 = getAllEarliestDates(searchCardsForMatches( re.compile(regex2) ))
plotDates(allDates2, regex2)
```

```python
# The ability "Horsemanship" was only ever printed in the "Portal Three Kingdoms" set from 1999.
regex2 = r'(.*)horsemanship(.*)'
allDates2 = getAllEarliestDates(searchCardsForMatches( re.compile(regex2) ))
plotDates(allDates2, regex2)
```

```python
# We can search for phrases of any length, like the phrase "without paying its mana cost".
# The huge spike in 2007 is from the mechanic "suspend" in the "Time Spiral" block.
regex3 = r'(.*)without paying its mana cost(.*)'
allDates3 = getAllEarliestDates(searchCardsForMatches( re.compile(regex3) ))
plotDates(allDates3, regex3)
```

```python
# Total number of new cards printed each year (as of October 4, 2021).
regex4 = r'.*'
allDates4 = getAllEarliestDates(searchCardsForMatches( re.compile(regex4) ))
plotDates(allDates4, regex4)
```
